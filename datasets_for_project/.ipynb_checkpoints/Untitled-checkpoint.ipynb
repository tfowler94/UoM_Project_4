```python
# Import Dependencies
import pandas as pd
import numpy as np
```

```python
# Load CSV files into new DataFrame
hc_expend_vs_gdp = "healthcare-expenditure-vs-gdp.csv"
hc_protection_cov = "health-protection-coverage.csv"
life_expect_vs_hc_expend = "life-expectancy-vs-health-expenditure.csv"
total_hc_expend_gdp = "total-healthcare-expenditure-gdp.csv"

# Read CSV file and store in Pandas DataFrame
hc_expend_vs_gdp_df = pd.read_csv(hc_expend_vs_gdp)
hc_protection_cov_df = pd.read_csv(hc_protection_cov)
life_expect_vs_hc_expend_df = pd.read_csv(life_expect_vs_hc_expend)
total_hc_expend_gdp_df = pd.read_csv(total_hc_expend_gdp)
```

```python
hc_expend_vs_gdp_df2 = hc_expend_vs_gdp_df.drop(columns="Continent")
hc_expend_vs_gdp_df3 = hc_expend_vs_gdp_df2.dropna(how="any")
hc_expend_vs_gdp_df3
```
