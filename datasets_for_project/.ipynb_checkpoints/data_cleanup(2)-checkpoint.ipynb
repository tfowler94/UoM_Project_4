```python
# Import Dependencies
import pandas as pd
import numpy as np
```

```python
# Load CSV files into new DataFrame
hc_expend_vs_gdp = "healthcare-expenditure-vs-gdp.csv"
hc_protection_cov = "health-protection-coverage.csv"
life_expect_vs_hc_expend = "life-expectancy-vs-health-expenditure.csv"
total_hc_expend_gdp = "total-healthcare-expenditure-gdp.csv"

# Read CSV file and store in Pandas DataFrame
hc_expend_vs_gdp_df = pd.read_csv(hc_expend_vs_gdp)
hc_protection_cov_df = pd.read_csv(hc_protection_cov)

life_expect_vs_hc_expend_df = pd.read_csv(life_expect_vs_hc_expend)
total_hc_expend_gdp_df = pd.read_csv(total_hc_expend_gdp)
```

```python
hc_expend_vs_gdp_df2 = hc_expend_vs_gdp_df.drop(columns="Continent")
hc_expend_vs_gdp_df3 = hc_expend_vs_gdp_df2.dropna(how="any")
hc_expend_vs_gdp_df4 = hc_expend_vs_gdp_df3.rename({"Year":"Year_1"}, axis='columns')

#cleaned dataframe
hc_expend_vs_gdp_df4.head()
```

```python
hc_protection_cov_df.head()
```

```python
hc_protection_cov_df2 = hc_protection_cov_df.drop(columns="Year")

#cleaned dataframe
hc_protection_cov_df2.columns
```

```python
life_expect_vs_hc_expend_df.head()
```

```python
life_expect_vs_hc_expend_df2 = life_expect_vs_hc_expend_df.drop(columns="Continent")
life_expect_vs_hc_expend_df3 = life_expect_vs_hc_expend_df2.dropna(how="any")
life_expect_vs_hc_expend_df4 = life_expect_vs_hc_expend_df3[life_expect_vs_hc_expend_df3['Life expectancy at birth, total (years)'].notna()]
life_expect_vs_hc_expend_df5 = life_expect_vs_hc_expend_df3.rename({"Year":"Year_2"}, axis='columns')

#cleaned dataframe
life_expect_vs_hc_expend_df5.columns
```

```python
total_hc_expend_gdp_df2 = total_hc_expend_gdp_df.rename({"Year":"Year_3"}, axis='columns')

#cleaned dataframe
total_hc_expend_gdp_df2.columns
```
